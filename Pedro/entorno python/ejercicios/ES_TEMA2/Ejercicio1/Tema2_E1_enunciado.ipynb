## Ejercicio 1: Lotería

Crea una **función llamada "premios"** para simular un sorteo de lotería de 2 cifras (del 10 al 99). Se harán 20 tiradas. Se debe utilizar el módulo **random** para generar las tiradas. Cada tirada es un numero, todas las tiradas se almacenan en un conjunto y se retornan. No puede haber 2 números iguales, es decir, si ya ha salido el 54, no puede volver a salir... Usad __conjuntos__

Crea otra **función llamada "apuesta"** con 5 tiradas aleatorias. Estas serán los números que juega el jugador. 

Crea otra función llamada **"comprobación"** que compara ambos conjuntos para saber cuántos números ha acertado. Debes hacer la comprobación mediante operaciones de conjuntos. Debes retornar una tupla con la cantidad de números acertados y la cantidad de números no acertados, es decir (acertados, no_acertados).

```python
import random

def premios():
    premios=set()
    while len(premios) < 20:
        premios.add(random.randint(10,99))

    return premios

def apuesta():
    tiradas=set()
    while len(tiradas) < 5:
        tiradas.add(random.randint(10,99))

    return tiradas

def comprobacion(premios,apuesta):
    acertados=len(premios.intersection(apuesta))
    no_acertados=len(premios.difference(apuesta))

    return acertados, no_acertados

num_premios=premios()
tiradas=apuesta()

num_acertados,num_no_acertados = comprobacion(num_premios,tiradas)
print(f"Numero acertados: {num_acertados}")
print(f"Numeros no acertados: {num_no_acertados}")

```

## Ejercicio 2. Agenda telefónica

Implementa una agenda telefónica simple utilizando un diccionario. Las claves serán los nombres y los valores serán tuplas con el número de teléfono y la dirección.

Debes crear las siguientes funciones:

+ Introducir un nuevo contacto
+ Buscar un contacto por nombre y mostrar su teléfono y dirección.
+ Eliminar contacto por nombre
+ Mostrar toda la agenda
+ Eliminar toda la agenda

```python
def introducir_contacto(agenda,nombre,telefono,direccion):
    if nombre and telefono and direccion:
        if nombre.lower() not in agenda.keys():
            agenda[nombre.lower()]=(telefono,direccion)
            return "Contacto añadido satisfactoriamente"
        else:
            return 'ERROR!:El contacto ya existe'
    else:
        return 'ERROR!:Rellene los campos correctamente por favor'
    
def buscar_contacto(agenda,nombre):
    if nombre in agenda.keys():
        return f" Contacto encontrado! Telefono: {agenda[nombre][0]} Direccion:{agenda[nombre][1]}"
    else:
        return "Contacto no encontrado"

def eliminar_contacto(agenda,nombre):
    if nombre.lower() in agenda.keys():
        agenda.pop(nombre.lower())
        return f"El contacto {nombre.title()} ha sido eliminado satisfactoriamente"
    else:
        return "Contacto no encontrado"
    
def mostrar_agenda(agenda):
    print(("-"*5)+"AGENDA"+("-"*5))
    for contacto,dato in agenda.items():
        print(f"{contacto.title()} numero: {dato[0]} dirección:{dato[1]}")

    print("-"*16)

def borrar_agenda(agenda):
    agenda.clear()

agenda=dict()
print(introducir_contacto(agenda,"Sergio","123456789","calle a"))
print(introducir_contacto(agenda,"Irene","123456789","calle a"))
print(introducir_contacto(agenda,"Paula","123456789","calle a"))
print(introducir_contacto(agenda,"Aitor","123456789","calle a"))
print(eliminar_contacto(agenda,"Irene"))
mostrar_agenda(agenda)
borrar_agenda(agenda)
mostrar_agenda(agenda)

```
