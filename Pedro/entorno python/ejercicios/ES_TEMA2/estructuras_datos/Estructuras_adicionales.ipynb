1.-DEQUE


```python
from collections import deque

historial = deque(maxlen=10)
revertidos = deque(maxlen=10)

def visitar_pagina(url):
    historial.append(url)
    revertidos.clear

def ir_atras():
    if(len(historial)>1):
        pag_actual=historial.pop()
        revertidos.appendleft(pag_actual)
        print(pag_actual)
    else:
        print('No hay paginas para volver atras')

def ir_adelante():
    if revertidos:
        pag_actual=revertidos.popleft()
        revertidos.append(pag_actual)
        print(pag_actual)
    else:
        print('No hay paginas para ir hacia delante')

visitar_pagina('https://hola.com')
ir_atras()
ir_adelante()
```

2. Counter

```python
from collections import Counter

texto = "Python es genial y PYTHON es práctico. Aprender python es útil"
texto=texto.lower().split()

counter = Counter(texto)
print(counter.most_common(5))
print(len(counter))
print(counter['python'])
```

3. namedtuple

```python
from collections import namedtuple
from collections import defaultdict

lista_alumnos=[]
#namedtuple
Alumno = namedtuple('Alumno',['nombre','grupo','nota'])

def insertar_alumno(tupla_alumno):
    alumn = Alumno(tupla_alumno[0],tupla_alumno[1],tupla_alumno[2])
    lista_alumnos.append(alumn)

def mayor_nota():
    print (max(lista_alumnos, key=lambda alumno: alumno.nota))

def media_grupos():
    d = defaultdict(list)
    for alumno in lista_alumnos:
        d[alumno.grupo].append(alumno.nota)

    medias = {}
    for grupo, notas in d.items():
        medias[grupo] = sum(notas) / len(notas)
        
    print(medias)


insertar_alumno(("Ana", "1A", 7.5))
insertar_alumno(("Luis", "1A", 6.0))
insertar_alumno(("Marta", "1B", 8.25))
insertar_alumno(("Juan", "1B", 9.0))


print(lista_alumnos)
mayor_nota()
media_grupos()
```
