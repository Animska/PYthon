```python
#GRAFICO DE LINEAS
import matplotlib.pyplot as plt
import numpy as np

# 1. Datos
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
ventas = [120, 145, 130, 180, 210, 195]

# 2. Configuración de la figura y el objeto Axes
fig, ax = plt.subplots(figsize=(8, 4), dpi=100)

# 3. Dibujo de la serie temporal
ax.plot(meses, ventas, marker='o', linestyle='-', color='#007acc', 
        linewidth=2.5, markersize=8, label='Ventas 2024')

# 4. Personalización del gráfico
ax.set_title('Evolución Mensual de Ventas', fontsize=14, fontweight='bold', loc='left')
ax.set_xlabel('Meses del año')
ax.set_ylabel('Ventas (Miles de €)')

# Añadir una anotación para resaltar el máximo
max_venta = max(ventas)
mes_max = meses[ventas.index(max_venta)]
ax.annotate(f'Pico: {max_venta}k', 
            xy=(mes_max, max_venta), 
            xytext=(mes_max, max_venta + 15),
            arrowprops=dict(facecolor='black', width=1, headwidth=5))

# Estética final
ax.grid(True, linestyle=':', alpha=0.6) # la rejilla de fondo
ax.set_ylim(100, 250) # Ajustamos límites para que no quede muy pegado arriba
ax.legend() # Mostramos la leyenda

plt.tight_layout()
plt.show()
```

```python
#NUBE DE PUNTOS
import matplotlib.pyplot as plt
import numpy as np

# 1. Preparación de datos
km_recorridos = np.random.uniform(8, 13, 50)
rendimiento = km_recorridos * 0.7 + np.random.normal(0, 1, 50)

# 2. Creación del objeto figura y el objeto axes
# Aunque sea solo uno, usamos subplots() sin argumentos (por defecto es 1x1)
fig, ax = plt.subplots(figsize=(8, 4))

# 3. Dibujo sobre el objeto 'ax'
ax.scatter(km_recorridos, rendimiento, color='red', alpha=0.6, edgecolors='white', label='Jugadores')

# 4. Personalización usando métodos del objeto 'ax'
ax.set_title('Relación: Distancia Recorrida vs Rendimiento', fontsize=14, fontweight='bold')
ax.set_xlabel('Kilómetros Recorridos')
ax.set_ylabel('Índice de Rendimiento')

# Añadir rejilla y leyenda
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend()

# 5. Mostrar 
plt.show()
```

```python
#GRAFICO DE BARRAS
import matplotlib.pyplot as plt

# 1. Datos
jugadores = ['Messi', 'Ronaldo', 'Mbappé', 'Haaland']
goles = [25, 22, 28, 32]

# 2. Creación de la figura y el objeto axes
fig, ax = plt.subplots(figsize=(8, 5), dpi=100)

# 3. Dibujo del gráfico de barras con atributos adicionales
# width: controla el ancho de la barra (0.8 es el valor por defecto)
# align: puede ser 'center' o 'edge'
barras = ax.bar(jugadores, goles, 
                color='skyblue',    # color del relleno
                edgecolor='navy',   # color del borde
                width=0.6,          # Baras un poco más delgadas 
                linewidth=1.5,      # Grosor del borde
                label='Goles 2024') # etiqueta para leyenda

# 4. Personalización del objeto 'ax'
ax.set_title('Goles por Jugador - Temporada Actual', fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('Cantidad de Goles', fontsize=12)
ax.set_ylim(0, 40) # Damos espacio arriba para las etiquetas

# Añadir etiquetas de valor sobre cada barra
# Iteramos sobre las barras creadas para poner el número exacto encima
ax.bar_label(barras, padding=3, fontsize=10, fontweight='bold')

# Estética y leyenda
ax.grid(True, axis='y', linestyle='--', alpha=0.4) # Rejilla solo horizontal
ax.legend()

plt.tight_layout()
plt.show()
```

```python
#BARRAS HORIZONTALES
import matplotlib.pyplot as plt

proyectos = ['Optimización de Infraestructura Local', 'Desarrollo de IA Generativa', 
             'Mantenimiento de Sistemas Legados', 'Ciberseguridad y Auditoría']
presupuesto = sorted([45000, 82000, 31000, 59000])

fig, ax = plt.subplots(figsize=(10, 5))

# Usamos barh en lugar de bar
ax.barh(proyectos, presupuesto, color='teal')

# Personalización
ax.set_title('Presupuesto por Proyecto (Q1 2026)', fontsize=14, fontweight='bold')
ax.set_xlabel('Presupuesto en Euros (€)')
ax.grid(True, axis='x', linestyle='--', alpha=0.6) # Rejilla solo en eje X

plt.tight_layout()
plt.show()
```

```python
tiendas = ['Tienda Norte', 'Tienda Sur', 'Tienda Este', 'Tienda Oeste']
hardware = [5000, 7000, 6000, 8000]
software = [3000, 4000, 2000, 5000]

fig, ax = plt.subplots(figsize=(9, 6))

# Dibujamos la primera capa
ax.bar(tiendas, hardware, label='Hardware', color='#3498db')

# Dibujamos la segunda capa encima (usando el parámetro bottom)
ax.bar(tiendas, software, bottom=hardware, label='Software', color='#e74c3c')

ax.set_title('Composición de Ventas por Sucursal', fontsize=14)
ax.set_ylabel('Ventas Totales (€)')
ax.legend()

plt.show()
```

```python
#GRAFICO DE TARTAS
import matplotlib.pyplot as plt

fuentes = ['Solar', 'Eólica', 'Gas', 'Hidroeléctrica']
cuotas = [15, 30, 45, 10]
colores = ['#ffda1f', '#87ceeb', '#ff7f50', '#1f77b4']

fig, ax = plt.subplots(figsize=(7, 7))

# autopct: muestra el porcentaje con un decimal
# startangle: gira el gráfico para mejor visualización
# explode: separa un trozo para destacarlo
ax.pie(cuotas, labels=fuentes, autopct='%1.1f%%', startangle=90, 
       colors=colores, explode=(0, 0, 0.1, 0), shadow=True)

ax.set_title('Mix Energético Mensual', fontsize=14, fontweight='bold')
plt.show()
```

```python
#HISTOGRAMA
import matplotlib.pyplot as plt
import numpy as np

# 1. Datos: Generamos 200 edades aleatorias entre 18 y 80 años
np.random.seed(1)
edades = np.random.randint(18, 81, size=200)

# 2. Creación de Figura y Axes
fig, ax = plt.subplots(figsize=(8, 4), dpi=100)

# 3. Dibujo del histograma
# bins=10: Agrupamos las edades en décadas o bloques de 6-7 años aproximadamente
ax.hist(edades, bins=10, color='wheat', edgecolor='saddlebrown', alpha=0.8)

# 4. Personalización del objeto 'ax'
ax.set_title('Distribución por Edades de la Muestra', fontsize=14, fontweight='bold')
ax.set_xlabel('Edad (Años)')
ax.set_ylabel('Número de Personas (Frecuencia)')

# Añadimos rejilla horizontal para facilitar la lectura de los conteos
ax.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
#GRAFICO HEXAGONOS
import matplotlib.pyplot as plt
import numpy as np

# Generamos 10.000 puntos aleatorios con distribución normal
x = np.random.normal(size=10000)
y = np.random.normal(size=10000)

fig, ax = plt.subplots(figsize=(8, 6))
grafica = ax.hexbin(x, y, gridsize=20) # gridsize: tamaño de los hexágonos
fig.colorbar(grafica, ax=ax, label='Conteo de puntos') # Añade la escala de color
ax.set_title('Densidad de Datos (Hexbin Plot)', fontsize=14)
plt.show()
```

```python
#BOXPLOT
dept_ventas = np.random.normal(2500, 500, 100)
dept_id = np.random.normal(3000, 200, 100)

fig, ax = plt.subplots(figsize=(8, 5))
ax.boxplot([dept_ventas, dept_id], tick_labels=['Ventas', 'I+D'])
ax.set_title('Análisis de Salarios por Departamento')
ax.set_ylabel('Euros (€)')
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
#GRAFICO DE BURBUJAS
import matplotlib.pyplot as plt
import numpy as np

# 1. Datos
presupuesto = [10, 25, 40, 55, 70, 85, 100] # Miles de €
ventas = [100, 150, 130, 250, 300, 310, 450] # Unidades vendidas
clientes = [50, 200, 800, 1200, 1500, 2500, 4000] # Nueva variable (Tamaño)

# 2. Creación de figura y axes
fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

# 3. Dibujo del gráfico de burbujas
# s: tamaño de la burbuja (usamos 'clientes' dividido por un factor para que no tape todo)
# c: color (podemos usar una variable o un color fijo)
# alpha: transparencia (VITAL para ver burbujas que se solapan)
burbujas = ax.scatter(presupuesto, ventas, 
                      s=np.array(clientes)/5, 
                      c=clientes, cmap='viridis', 
                      alpha=0.6, edgecolors="grey", linewidth=1)

# 4. Personalización
ax.set_title('Impacto de Marketing en Ventas y Captación', fontsize=14, fontweight='bold')
ax.set_xlabel('Presupuesto de Marketing (k€)')
ax.set_ylabel('Ventas (Unidades)')

# Añadimos una barra de color para la variable 'clientes'
fig.colorbar(burbujas, ax=ax, label='Nº de Clientes')

ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
#MATRIZ SUBPLOTS 2X2
import matplotlib.pyplot as plt
import numpy as np

# Datos sintéticos
x = np.linspace(0, 10, 100)

# 1. Creamos la cuadrícula (2 filas y 2 columnas)
# sharex=True permite que todos compartan el mismo eje X si los datos son comparables
fig, ax = plt.subplots(2, 2, figsize=(12, 8), dpi=100)

# 2. Gráfico [0, 0] - Superior Izquierda (Líneas)
ax[0, 0].plot(x, np.sin(x), color='blue')
ax[0, 0].set_title('Seno de X')
ax[0, 0].grid(True, alpha=0.3)

# 3. Gráfico [0, 1] - Superior Derecha (Dispersión)
ax[0, 1].scatter(x, np.cos(x) + np.random.normal(0, 0.1, 100), color='red', s=10)
ax[0, 1].set_title('Coseno con Ruido')

# 4. Gráfico [1, 0] - Inferior Izquierda (Barras)
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 2, 5]
ax[1, 0].bar(categorias, valores, color='green')
ax[1, 0].set_title('Frecuencias por Categoría')

# 5. Gráfico [1, 1] - Inferior Derecha (Histograma)
datos_hist = np.random.randn(1000)
ax[1, 1].hist(datos_hist, bins=30, color='orange', edgecolor='black')
ax[1, 1].set_title('Distribución Normal')

# --- AJUSTES FINALES ---
plt.tight_layout() #evita que los títulos se solapen con los ejes
plt.suptitle('Panel de Control de Análisis Exploratorio', fontsize=16, y=1.02) # Título global
plt.show()
```

```python
#grafico de linea con 3 lineas
import matplotlib.pyplot as plt

# 1. Datos
partidos = ['P1', 'P2', 'P3', 'P4', 'P5']
jugador_A = [10.5, 11.2, 9.8, 12.1, 10.9]
jugador_B = [9.2, 9.5, 10.1, 10.0, 9.8]
jugador_C = [11.8, 12.5, 12.2, 11.5, 12.0]

# 2. Creación de Figura y Axes
fig, ax = plt.subplots(figsize=(10, 6), dpi=100)

# 3. Dibujo de las 3 líneas
# Usamos diferentes colores, estilos de línea y marcadores para máxima claridad
ax.plot(partidos, jugador_A, label='Jugador A', color='blue', marker='o', linewidth=2)
ax.plot(partidos, jugador_B, label='Jugador B', color='green', marker='s', linestyle='--', linewidth=2)
ax.plot(partidos, jugador_C, label='Jugador C', color='red', marker='^', linestyle=':', linewidth=2)

# 4. Personalización del objeto 'ax'
ax.set_title('Comparativa de Rendimiento Físico (Distancia)', fontsize=14, fontweight='bold')
ax.set_xlabel('Jornada de Liga')
ax.set_ylabel('Kilómetros Recorridos')

# Añadir una zona sombreada para resaltar el "rango de fatiga" 
ax.axhspan(11.5, 13, color='grey', alpha=0.1, label='Zona de Alto Desgaste')

# Estética final
ax.grid(True, linestyle=':', alpha=0.6)
ax.legend(loc='upper left', frameon=True, shadow=True)

plt.tight_layout()
plt.show()
```
