```python
#importar csvs y datasets
import pandas as pd

data_embarazadas= pd.read_csv("embarazo_adolescente.csv")
data_malaria= pd.read_csv("Malaria.csv")
```

```python
data_embarazadas.drop(columns = ['Indicator'], inplace=True)
data_embarazadas.columns = ['Pais', 'Anio', 'Embarazos adolescentes']
data_embarazadas.head(5)
```

```python
data_malaria.drop(columns = ['Indicator'], inplace=True)
data_malaria.columns = ['Pais', 'Anio', 'Casos de Malaria']
data_malaria.head(5)
```

```python
finaldf= pd.merge(data_malaria,data_embarazadas, how='inner', on=['Pais', 'Anio'])
finaldf.head()
```

```python
finaldf.query('Anio == 2018').head()
```

```python
finaldf.groupby('Pais')['Casos de Malaria'].max().sort_values(ascending=False).head(10)
```

```python
finaldf.groupby('Pais')['Casos de Malaria'].mean().sort_values(ascending=False).head(10)
```
