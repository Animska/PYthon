```python
import pandas
from influxdb import InfluxDBClient

HOST = '192.168.1.8'
PORT = 8886
USUARIO = 'Pythonread'
PASSWORD = 'read1234'
DATABASE = 'agro_sensex'

cliente = InfluxDBClient(HOST,PORT,USUARIO,PASSWORD)

print(cliente.ping())
```

```python
#Lista de base de datos en forma de diccionario con clave 'name'
print(cliente.get_list_database)
```

```python
#Seleccionar base de datos
cliente.switch_database(DATABASE)
```

```python
#Mostrar measurements(devuelve un JSON que se convierte a un set de  listas de diccionarios)
resultado = cliente.query('SHOW MEASUREMENTS')
for point in resultado.get_points():
    print(point)
```

```python
#Mostrar tag keys de la tabla '%'
resultado = cliente.query('SHOW TAG KEYS FROM "%"')
for point in resultado.get_points():
    print(point)
```

```python
#mostrar valores de la tag key 'domain'
resultado = cliente.query('SHOW TAG VALUES FROM "%" WITH KEY = "domain"')
for point in resultado.get_points():
    print(point)
```

```python
resultado = cliente.query('SHOW FIELD KEYS FROM "%"')
```

```python
resultado = cliente.query(
                        f'SELECT value FROM ""%'
                        f'WHERE "entity_id" = \'aqsex_sdf_huerto_lht65n_humedad\''
                        f'AND time > now() - 1h'
                        )

for point in resultado.get_points:
    print(point)
```

```python
#pasar influxdb a dataframe
resultado = cliente.query(
                        f'SELECT value FROM ""%'
                        f'WHERE "entity_id" = \'aqsex_sdf_huerto_lht65n_humedad\''
                        )
df = pandas.DataFrame(resultado.get_points())
df.head(5)
```

```python
#pasar influxdb a series
resultado = cliente.query(
                        f'SELECT value FROM ""%'
                        f'WHERE "entity_id" = \'aqsex_sdf_huerto_lht65n_humedad\''
                        )
df = pandas.DataFrame(list(resultado.get_points()))
df['time'] = pandas.to_datetime(df['time'])
df = df.set_index['time']
humedad = df['value']
humedad.name = "Humedad en %"
humedad.head(5)
```

```python
#indexar por fecha [a√±o-mes-dia-hora-minuto-segundo]
humedad.loc['2025-12']
#se puede slicear
humedad.loc['2025-12-1' : '2025-12-4']
```

```python
#Deltatime
mascara = humedad.index.max() - pandas.Timedelta('5h')
humedad.loc[mascara:]
```
