```python
#Preparacion
import pandas as pandas

#Humedad
#cargar los archivos en un dataframe
df_humedad = pandas.read_csv('humedad.csv',sep=",",header=0)
#Indicar que la columna time es de tipo datetime
df_humedad['time'] = pandas.to_datetime(df_humedad['time'])
#Pasar la columna time a etiqueta de las filas
df_humedad = df_humedad.set_index('time')
# Indexar la columa value para obtener solamente una Series.
humedad = df_humedad['value']
humedad.name = "Humedad en %"
humedad.head(5)


#Temperatura
df_temperatura = pandas.read_csv('temperatura.csv',sep=",",header=0)
df_temperatura['time'] = pandas.to_datetime(df_temperatura['time'])
df_temperatura = df_temperatura.set_index('time')
temperatura = df_temperatura['value']
temperatura.name = "Temperatura en Cº"
temperatura.head(5)
```

```python
#1.-Consultar la humedad del día de ayer, todas las filas
humedad.loc['2025-12-2']
```

```python
#2.-Consultar la media, mínima y máxima, y std del día de ayer, humedad y temperatura
print("-HUMEDAD")
print(f"Humedad media: {humedad.loc['2025-12-2'].mean():.2f}%")
print(f"Humedad maxima: {humedad.loc['2025-12-2'].max():.2f}%")
print(f"Humedad minima: {humedad.loc['2025-12-2'].min():.2f}%")
print(f"Humedad desviacion tipica: {humedad.loc['2025-12-2'].std():.2f}%")

print()

print("-TEMPERATURA-")
print(f"Temperatura media: {temperatura.loc['2025-12-2'].mean():.2f}Cº")
print(f"Temperatura maxima: {temperatura.loc['2025-12-2'].max():.2f}Cº")
print(f"Temperatura minima: {temperatura.loc['2025-12-2'].min():.2f}Cº")
print(f"Temperatura desviacion tipica: {temperatura.loc['2025-12-2'].std():.2f}Cº")

```
