Del conjunto de datos del titanic.csv realiza las siguientes consultas usando query y sin usar query, generalmente con loc bastará. 

En algunas consultas hay que usar más cosas aparte del query. 

```python
import pandas

df = pandas.read_csv('titanic.csv')
df.head()
```

```python
# Consultar de los hombres, las columnas survived, pclass, age

#pandas
df_hombres = df[df['Sex'] == 'male'][['Survived', 'Pclass', 'Age']]
df_hombres

#query
df.query("Sex == 'male'")[['Survived', 'Pclass', 'Age']]
```

```python
# Pasajeros de clase menor a 3 (2 y 1)

#pandas
df_clase=df[df['Pclass'] < 3]
df_clase

#query
df.query(' Pclass < 3')
```

```python
# Hombres supervivientes, 2 consultas con operador and

#pandas
df_supervivientes = df[(df['Sex'] == 'male') & (df['Survived'] == 1)]
df_supervivientes

#query
df.query('Sex == "male" and Survived == 1')
```

```python
# Pasajeros donde Hermanos/Conyuges es mayor a padres/hijos

#pandas
df_hermanos = df[df['SibSp'] > df['Parch']]
df_hermanos

#query
df.query('SibSp > Parch')
```

```python
# Pasajeros de Embarked S o C

#pandas
df_embarked=df[(df['Embarked'] == 'S') | (df['Embarked'] == 'C')]
df_embarked

#query
df.query('Embarked == "S" or Embarked == "C"')
```

```python
# Cuantos pasarejos existen del Embarked S, hazlo a través de una variable y un f-string

#pandas
embarked="S"
df_embarked_s=df[df['Embarked'] == embarked].shape[0]
print(f"Existen {df_embarked_s} pasajeros en Embarked S")

#querys
embarked="S"
pasajeros_s = df.query(f"Embarked == '{embarked}'").shape[0]
print(f"Existen {pasajeros_s} pasajeros en Embarked S")
```

```python
# Sobrevivientes mayores de edad 

#pandas
df_supervivientes_mayores = df[(df['Age'] >= 18) & (df['Survived'] == 1)]
df_supervivientes_mayores

#query
df.query('Age >= 18 and Survived == 1')
```

```python
# Cuantos no supervivientes menores de edad habia de clase 3

#pandas
supervivientes_menores_clase_3=df[(df['Age'] < 18) & (df['Survived'] == 0) & (df['Pclass'] == 3)].shape[0]
print(f"Habia {supervivientes_menores_clase_3} no supervivientes menores de edad en 3º clase")

#query
supervivientes_menores_clase_3=df.query('Age < 18 and Survived == 0 and Pclass == 3').shape[0]
print(f"Habia {supervivientes_menores_clase_3} no supervivientes menores de edad en 3º clase")
```

```python
# La edad media de los supervivientes mujeres de clase 1 o de clase 2

#pandas
df_mujeres_clase_12= df[(df['Sex'] == 'female') & (df['Survived']==1) & (df['Pclass'].isin([1,2]))]
media_edad = df_mujeres_clase_12['Age'].mean()
print(f"La media de edad de las mujeres supervivientes de 1º o 2º clase es {media_edad} años")

#query
mujeres_clase_12 = df.query(f"Sex == 'female' and Survived == 1 and Pclass in [1,2]")
media_edad = mujeres_clase_12['Age'].mean()
print(f"La media de edad de las mujeres supervivientes de 1º o 2º clase es {media_edad} años")

```

```python
# Mostrar supervivientes donde la edad es desconocida

#pandas
df_supervivientes_edad_null=df[(df['Survived'] == 1) & (df['Age'].isna())]
df_supervivientes_edad_null

#query
df.query('Survived == 1 and Age.isna()')
```

```python
# Mostrar aquellos que no sobrevivieron y pagaron un precio entre 15 y 20

#pandas
df_survivors_15_20 = df[(df['Survived'] == 0) & (df['Fare'] > 15) & (df['Fare'] < 20)]
df_survivors_15_20

#querys
df.query('Survived == 0 and Fare > 15 and Fare <20')
```

```python
# Consultar a través de una lista los los valores donde Embarked es S y C
lista=["S","C"]
#pandas
df_embarked=df[df['Embarked'].isin(lista)]
df_embarked

#query
df.query(f'Embarked in {lista}')
```

```python
# Consultar aquellos donde el número de familiares totales es mayor a 5

#pandas
df_familiares_mayor_5= df[df['SibSp']+df['Parch'] > 5]
df_familiares_mayor_5

#querys
df.query('Parch +SibSp > 5')
```

```python
# Consultar cuántos supervivientes viajab sin familiares

#pandas
supervivientes_sin_familiares=df[(df['SibSp']+df['Parch'] <=0) & (df['Survived']==1)].shape[0]
print(f"{supervivientes_sin_familiares} supervivientes viajaban sin familiares")

#query
supervivientes_sin_familiares=df.query('Parch + SibSp <=0 and Survived == 1').shape[0]
print(f"{supervivientes_sin_familiares} supervivientes viajaban sin familiares")
```

```python
# Consultar la edad mínima del hombre superviviente de tercera clase

#pandas
edad_minima_hombre_superviviente_3clase = df[(df['Sex'] == 'male') & (df['Survived'] == 1) & (df['Pclass'] == 3)]['Age'].min()
print(f"La edad mínima del hombre superviviente de tercera clase es {edad_minima_hombre_superviviente_3clase}")

#query
edad_minima_hombre_superviviente_3clase=df.query('Sex == "male" and Survived == 1 and Pclass == 3')['Age'].min()
print(f"La edad mínima del hombre superviviente de tercera clase es {edad_minima_hombre_superviviente_3clase}")
```

```python
# Meter en una lista la cantidad de pasajeros por tipo de Embarked

#pandas
dict_embarked = dict()
lista_embarked=df['Embarked'].unique()

for tipo in lista_embarked:
    dict_embarked[tipo] = df[(df['Embarked'] == tipo)].shape[0]

print(dict_embarked)

#query
dict_embarked = dict()
lista_embarked=df['Embarked'].unique()

for tipo in lista_embarked:
    dict_embarked[tipo] = df.query(f'Embarked == "{tipo}"').shape[0]

print(dict_embarked)
```
