# Ejercicio 1

+ Tenemos una lista que contiene elementos de varios tipos.
+ Necesitamos hacer un programa que nos divida la lista original en 3 listas, donde cada una contiene solamente los elementos de un tipo (int, float, str) ignorando los elementos del resto de tipos.
+ Cuando se encuentre un elemento que no sea int, float, str, imprimir un mensaje "Elemento no válido -> elemento"
+ Imprimir las 3 listas para comprobar el resultado.


Lista inicial ['h', 2, 3.5, 5, [9],  'j', 4, 7.8, 'o']

Resultado:

+ lista_int = [2, 5, 4]
+ lista_float = [3.5, 7.8]
+ lista_str = ['h', 'j', 'o']


# Ayuda

La función **isinstance()** en Python sirve para determinar si un objeto es una instancia de una clase específica. 

Sintaxis: isinstance(objeto, clase)

```python
lista_inicial=['h', 2, 3.5, 5, [9],  'j', 4, 7.8, 'o']
lista_int=[]
lista_float=[]
lista_str=[]

for elemento in lista_inicial:
    if isinstance(elemento,int):
        lista_int.append(elemento)
    elif isinstance(elemento,float):
        lista_float.append(elemento)
    elif isinstance(elemento,str):
        lista_str.append(elemento)  
    else:
        print(f"Elemento no valido -> {elemento}")

print(lista_int)
print(lista_float)
print(lista_str)


```

# Ejercicio 2: Agenda de contactos

+ Crea una agenda de contactos desarollando varias funciones. 
+ La agenda se compone de 2 listas, una lista de nombres y una lista de números de teléfonos. Ambas listas se relacionan por el índice, es decir, si Juán ocupa el índice 5 en la lista de nombres, su número será el índice 5 de la lista de números.
+ Crear las siguientes funciones para manejar la lista:
    + **Insertar contacto**. Se debe pedir al usuario, dentro de la función, que introduzca el nombre y el número. Si el nombre ya existe, el usuario debe tener 3 opciones: actualizar el número; introduzca un nombre diferente; ignorar la operación. La agenda en todo momento debe estar orenada por nombre alfabéticamente, por lo que cada nuevo contacto se debe insertar en su posición correcta. 
    + **Mostrar agenda**.
    + **Buscar contacto**: Se debe permitir buscar por nombre o por número de teléfono. Y mostrar el otro dato, es decir, si se busca por número, mostrar el nombre a quién pertenece.
    + **Eliminar contacto**: Preguntar al usuario que nombre desea eliminar (no olvidar borrar también el número). Si no existe el contacto se debe informar de ello.
    + **Borrar toda la agenda**

```python
nombres=[]
numeros=[]

def insertar_alfabeticamente(nombre,numero):
    for indice,nombre_agenda in enumerate(nombres):
        if nombre < nombre_agenda:
            nombres.insert(indice,nombre)
            numeros.insert(indice,numero)
            print(f"Contacto {nombre} añadido satisfactoriamente")
            return
    nombres.append(nombre)
    numeros.append(numero)
    print(f"Contacto {nombre} añadido satisfactoriamente")

def insertar_contacto():
    try:
        nombre=input("introduzca el nombre").lower()
        numero=int(input("introduce el numero de telefono"))
        if nombre in nombres:
            try:
                operacion=int(input("""
                    ERROR! Ese nombre ya existe en la agenda.
                    Seleccione una de estas opciones:
                    0.-Introducir un nombre diferente.
                    1.-Actualiza el numero de telefono
                    2.-Ignora la operacion
                """))
                match operacion:
                    case 0:
                        while nombre in nombres:
                            nombre=input("introduzca el nombre de nuevo").lower()
                        insertar_alfabeticamente(nombre,numero)
                    case 1:
                        indice= nombres.index(nombre)
                        numeros[indice]=numero
                        print(f"EL numero del Contacto {nombre} ha sido actualizado satisfactoriamente")
                    case 2:
                        pass
                    case _:
                        print("Operacion incorrecta")
            except ValueError:
                print("ERROR! La operacion debe ser un numero")
        else:
            insertar_alfabeticamente(nombre,numero)
    except ValueError:
        print("numero de telefono no valido")

def mostrar_agenda():

    if nombres:
        for index,nombre in enumerate(nombres):
            print(f"Nombre: {nombre.title()} - Numero: {numeros[index]}")
    else:
        print("Agenda Vacia")

def buscar_contacto():
    termino_busqueda=input("introduce el nombre o el numero de telefono que desee buscar").lower()
    if termino_busqueda in nombres:
        print(f"Buscando el numero de: {termino_busqueda.title()}...")
        indice= nombres.index(termino_busqueda)
        print(f"Su numero de telefono es {numeros[indice]}")
    elif termino_busqueda.isdigit() and int(termino_busqueda) in numeros:
        print(f"Buscando el contacto de: {termino_busqueda}...")
        indice= numeros.index(int(termino_busqueda))
        print(f"Su nombre de contacto es {nombres[indice].title()}")
    else:
        print("No se ha encontrado nada.")

def borrar_contacto():
    termino_busqueda=input("introduce el nombre del contacto que desea eliminar").lower()
    if termino_busqueda in nombres:
        indice= nombres.index(termino_busqueda)
        nombres.pop(indice)
        numeros.pop(indice)
        print(f"Se ha borrado al usuario {termino_busqueda.title()} satisfactoriamente")
    else:
        print("Usuario no encontrado en agenda")

def borrar_agenda():
    nombres.clear()
    numeros.clear()




opcion=0
while opcion!=5:
    try:
        opcion=int(input("""
                    Bienvenido a tu agenda,¿Que operacion deseas realizar?:
                    (introduzca un numero para seleccionar la opcion)
                    0.-Añadir un Contacto.
                    1.-Mostrar la Agenda
                    2.-Buscar un Contacto
                    3.-Eliminar un Contacto
                    4.-Vaciar la Agenda
                    5.-Salir
                """))
        match opcion:
            case 0:
                insertar_contacto()
            case 1:
                mostrar_agenda()
            case 2:
                buscar_contacto()
            case 3:
                borrar_contacto()
            case 4:
                borrar_agenda()
            case 5:
                print("¡Hasta la proxima!")
            case _:
                print("ERROR!Opcion incorrecta, debe estar entre 0 y 5")
    except ValueError:
        print("ERROR!La opcion debe ser un numero")
    
    print("--------------------------------------------------------------------------------------")
```
