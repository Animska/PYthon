# Ejercicio 1: NumPy (6 puntos)

Cada celda es 0.5 puntos a excepción de la última que vale 1 punto.

La siguiente celda solamente contiene los __datos__: 3 ndarrays con notas de alumnos y un ndarray adicional con los nombres.

```python
import numpy as np
# ======
# DATOS
#=======
nota_mates = np.array([4.3, 9.1, 7.2 , 4.8, 7.2, 4.4, 5.5, 4.2, 6.9, 3.3, 4.5, 5.6, 5.4, 6.8, 5.1])
nota_lengua = np.array([7.1, 8.7, 6.5 , 7.5, 5.6, 6.2, 5.4, 4.8 , 3.6, 4.8, 6.4, 5.6, 5.1 , 8.4, 5.0]) 
nota_fisica = np.array([7.4, 10, 7.8 , 5.4, 8.4, 3.2, 5.8, 6.4, 7.3, 3.5, 3.4, 5.6, 6.2, 4.1, 5.2])
alumnos = np.array(["Juan", "Noelia", "Pedro", "Marta", "Antonio", "Carlos", "Patricia", "Carmen", "Manuel", "Jorge",
                "Jose", "Sara", "María", "Rubén", "Bárbara"])
```

```python
#1.1
# Crea un ndarray del mismo tamaño y tipo de datos que notas_mates llamado "nuevo" con los valores a 1
nuevo=np.ones_like(nota_mates)

# Suma notas_mates y "nuevo" de forma vectorizada y guarda el resultado en ndarray "suma". Imprime el resultado.
suma=np.add(nota_mates,nuevo)

# Muestra el número de notas de suma
suma
```

```python
#1.2
# Muestra si existe al menos un alumno que haya suspendido lengua. Con mostrar True o False es sufiente.
print(np.any(nota_lengua<5))
```

```python
#1.3
# Muestra el nombre de los alumnos que han aprobado matemáticas
arr_aprobados_lengua = (nota_mates >= 5)
#esto devuelve un array de true y falses
print(alumnos[arr_aprobados_lengua])
```

```python
#1.4
# Muestra el nombre de los alumnos que han sacado, en matemáticas, menor nota que la media de matemáticas
media_mates=np.mean(nota_mates)
arr_menor_notamedia_mates = (nota_mates < media_mates)
print(alumnos[arr_menor_notamedia_mates])
```

```python
#1.5
#Muestra el nombre de los alumnos que han sacado mayor nota en mates que en física
#nombro el array segun el ejercicio para que no ocupe tanto espacio
arr_1_5 = (nota_mates > nota_fisica)
print(alumnos[arr_1_5])
```

```python
#1.6
# Muestra el nombre de los alumnos que han suspendido matemáticas y lengua
arr_1_6 = (nota_mates < 5) & (nota_lengua < 5)
print(alumnos[arr_1_6])
```

```python
#1.7
# Muestra el nombre de los alumnos que han aprobado todo
arr_1_7 = (nota_mates >= 5) & (nota_lengua >= 5) & (nota_fisica >= 5)
print(alumnos[arr_1_6])
```

```python
#1.8
# Genera un ndarray de strings, indicando en cada valor "lengua" o "fisica"
# El ndarray debe indicar, por cada alumno, la asignatura donde tiene la nota más alta entre lengua y fisica
arr_1_8=np.empty_like(alumnos)

for index,alumno in enumerate(alumnos):
    if nota_lengua[index] < nota_fisica[index]:
        arr_1_8[index] = "Lengua"
    elif nota_lengua[index] > nota_fisica[index]:
        arr_1_8[index] = "Fisica"
    else:
        arr_1_8[index] = "Empate"

print(arr_1_8)
```

```python
#1.9
# Genera un array que guarde la nota media de las tres asignaturas, redondeando a 2 decimales. Una el método ndarray.round() para redondear.
arr_1_9=np.empty_like(nota_lengua)

for index,alumno in enumerate(alumnos):
    media= np.round(np.mean((nota_lengua[index],nota_mates[index],nota_fisica[index])),2)
    arr_1_9[index] = media

print(arr_1_9)
```

```python
#1.10
# Calcula y muestra cuantos alumnos han aprobado y cuantos han suspendido en matemáticas (se aprueba con 5 o más)

print(f"aprobados: {len(alumnos[nota_mates>=5])}")
print(f"suspendidos: {len(alumnos[nota_mates<5])}")
```

```python
#1.11
# Calcula y muestra un array llamado "maximas" que guarde la nota más alta de cada alumno entre las 3 asignaturas
maximas=np.empty_like(nota_lengua)
for index,alumno in enumerate(alumnos):
    maxima= max((nota_lengua[index],nota_mates[index],nota_fisica[index]))
    maximas[index] = maxima

print(maximas)

# Muestra el nombre del alumno que ha sacado la nota más alta entre todas las asignaturas
nota_maxima = np.max(maximas)
for indice,alumno in enumerate(alumnos):
    if maximas[indice] == nota_maxima:
        print(alumno)

```
