# EJERCICIO 1

+ Pide al usuario una palabra y comprueba si la palabra es un palíndromo. (no hace falta controlar las tildes)
+ Debemos asegurarnos de que se introduzca solo 1 palabra sin signos de puntuación ("ho-la" o "hola?" no son palabras correctas). Podemos usar el método isalpha.
+ Si no es una palabra simple mostramos un error.
+ El programa no es case-sensitive

```python
palabra = input("Introduce una palabra: ").lower()
if palabra.isalpha():
    if palabra == palabra[::-1]:
        print(f"{palabra} es un palíndromo")
    else:
        print("No es un palíndromo")
else:
    print(f"La palabra {palabra} no es correcta")
```

# EJERCICIO 2

+ Disponemos de un string que contiene una serie de frutas separadas por punto y coma ';' frutas="manzana;albaricoque;pera;naranja;kiwi;uva"
+ Divide el string utilizando split y el caracter separador indicado
+ Recorre la lista resultante con un for
+ Muestra solamente las frutas que comienzan por vocal  (mayúsculas o minúsculas)

```python
frutas="manzana;albaricoque;pera;naranja;kiwi;uva"
#Divide el string en una lista utilizando el separador ';'
lista_frutas = frutas.split(";")

print("Frutas que comienzan por vocal:")
for fruta in lista_frutas:
    if fruta[0].lower() in "aeiou":
        print(fruta)
```

# EJERCICIO 3: VALIDACIÓN SIMPLE SIN USAR REGEX

+ Pide al usuario que introduzca un correo electrónico y verifica si contiene un formato adecuado.

+ El formato a comprobar (simplificado) será nombre@dominio.es o nombre@dominio.com:

    + El nombre puede ser cualquier cadena sin @
    + El dominio puede ser cualquier cadena sin @ ni .
    + Debe acabar obligatoriamente en .com o .es. 
    + Es decir: 
        + Solo puede haber un @ en toda la cadena. 
        + A la derecha de la @ solo puede haber 1 punto que sirve para separar el dominio.

+ A cada tipo de error mostrar un mensaje personalizado que informe al usuario
+ No es case-sensitive
+ Métodos a usar: lower, endswith, split.

Se proponen unos if para organizar el código:
+ Comprobar que solo tenga 1 @ y que acaba en .com o .es. Si no lo cumple mostrar "El formato debe ser..."
+ Comprobar que el nombre no esté vacío. Si no lo cumple mostrar "El nombre está vacío"
+ Comprobar que después del @ solo tiene 1 punto. Si tiene más mostrar el error.
+ Comprobar que dominio no esté vacío.

Si el correo es correcto mostrar el mensaje.


```python
# Solicitar el correo electrónico al usuario
correo = input("Introduce un correo electrónico: ").lower()

partes = correo.split('@')
# Verificamos que tiene solo 1 @; que acaba en .com o .es; que no tiene espacios
if (len(partes) != 2) or (not correo.endswith((".com", ".es"))) or (" " in correo):
    print("El formato es nombre@dominio.es/com")
else:
    nombre, dominio = partes
    # Verificar que 1º parte no esté vacío
    if not nombre:
        print("El nombre está vacío")
    else:
        # Verificar que 2º solo 1 un punto
        partes_dominio = dominio.split('.')
        if len(partes_dominio) != 2:
            print("Solo puede haber 1 punto en la segunda parte")
        else:
            dom1, dom2 = partes_dominio
            # Verificar que subdominio no esté vacío
            if not dom1:
                    print(f"El subdominio está vacío")
            else:
                print("Correo válido")
```

## Mismo ejercicio con expresiones reguales

Usa la librería __re__ para crear una expresión regular y validad el correo.
En este caso no se muestran mensajes personalizados, hazlo en un paso solamente con 1 expresión regular. 

https://docs.python.org/es/3.14/library/re.html


```python
import re

correo = input("Introduce un correo electrónico: ").lower()

# -la r del principio significa que es una cadena cruda, no se aplican los caracteres escapados
# -^ indica el inicio de la cadena.
# -[^@\s]+ significa uno o más caracteres que no sean ni @ ni espacios. Esto corresponde a la parte local del correo (antes de la arroba).
# -@ es el carácter obligatorio @ que separa usuario y dominio.
# -[^@.\s]+ es uno o más caracteres que no son @, ni punto . ni espacios, para la parte del dominio (antes del punto final).
# -\. es el carácter literal punto . que separa el dominio y la extensión.
# -(com|es) indica que la extensión debe ser o "com" o "es".
# -$ indica el final de la cadena.
patron = r"^[^@\s]+@[^@.\s]+\.(com|es)$"

# Validar el correo con `re.match()`
if re.match(patron, correo):
    print(f"El correo '{correo}' tiene un formato válido.")
else:
    print(f"El correo '{correo}' no tiene un formato válido.")
```

# EJERCICIO 4

+ Inicialmente le pedimos al usuario que introduzca una frase. Prueba con la frase "¿hola, qué tal? Bie?venido Python..."
+ Limpia la frase para obtener una lista de palabras homogénea (minúsculas, eliminando signos de puntuación al inicio y al final de cada palabra, filtrando las mal formadas) lista=[hola, qué, tal, python]
+ Pide al usuario que introduza una palabra y verifica si la palabra está presente en la frase.
+ Debe ser una palabra correcta (solo debe tener caracteres alfabéticos). 
+ Si el usuario introduce una palabra errónea se le mostrará error.
+ El programa no es case-sensitive
+ Métodos a usar: lower, isalpha, split, strip, los signos de puntuación están en librería string, constante punctuation (¿ ¡ no están incluidos en la constante).

```python
import string

palabras = input("Introduce una frase: ").lower().split()

filtradas = []
for index, palabra in enumerate(palabras):
    palabra = palabra.strip(string.punctuation + "¿¡")
    if palabra.isalpha():
        filtradas.append(palabra)
    else:
        print("Palabra eliminada:", palabra)
print("Lista de palabras filtrada: ", filtradas)

#Buscar palabra
palabra = input("Introduce una palabra a buscar: ").lower()
if palabra.isalpha():
    if palabra in filtradas:
        print(f"La palabra '{palabra}' existe en la frase")
    else:
        print(f"La palabra '{palabra}' no existe en la frase")
else:
    print(f"La palabra no es correcta -> {palabra}")
```
